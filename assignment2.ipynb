```python
#2 list wrangling:
many_nums=[]
l3=[]             # contains elements that are divisable by 3

for i in range(0,101):
    many_nums.append(i)
    print(many_nums)
    
for i in many_nums:        #creating new list to store divisable by 3
    if(i%3==0):
        l3.append(i)
        print(l3)
for i in many_nums[:]:      #list only with even nums
    if i%2!= 0:
        many_nums.remove(i)
print(many_nums)

del many_nums [0:3]
del many_nums[-3:]
print("the sum is:",sum(many_nums))
```

```python
#3 pattern loop design:
row=11

for i in range (0,row):                #loop to keep the design going      
    for j in range(0,i +1):            #inner loop for each column to create the design
        print("*", end= ' ')
    print("\r")
for i in range (row, 0, -1):              #reverse for bottom portion
    for j in range(0,i -1):
        print("*", end= ' ')
    print("\r")
```

```python
#4 palindrome return true or false
def palindrome(word):
    reversed_word = word[::-1]
    status=1
    if(word!=reversed_word):
        status=0
    return status

word = input("Enter the string: ").lower()
status=palindrome(word)
if(status):
    print("True")
else:
    print("False")
```

```python
#5 dictionary practice

NASA= {
    3:'Gemini' ,
    13: 'Apollo',
    5: 'Mercury Redstone'
        
}
print(NASA)
NASA[4] = 'Redstone'
NASA[16] = 'Project x'  #adding key and value
print(NASA)

NASA[16]= 'project y'         #altering value
print(NASA)

del NASA[16]
print(NASA)

for k,v in NASA.items():           #loop and print out
    print(str(k)+ ":"+v)

for k in NASA:         # printing only key values from dictionary
    print(k)
```
